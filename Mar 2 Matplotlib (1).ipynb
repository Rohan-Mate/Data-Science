```python
x = ["a","b","c","d","e"]
y = np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.bar(x,y)
# plot the graph - categorical vs numerical
```

```python
plt.figure(figsize = (6,2))
plt.bar(x,y , color = "#C728C7")
plt.xlabel(" representing my cat data ")
plt.ylabel(" representing my num data ")
plt.title(" this is my graph plot")
plt.grid()
```

```python
plt.figure(figsize = (6,2))
plt.barh(x,y , color = "#C728C7")
plt.xlabel(" representing my cat data ")
plt.ylabel(" representing my num data ")
plt.title(" this is my graph plot")
plt.grid()
# transpose
```

```python
data = [1,2,3,4,5,6,6,6,7,7,8,8,8,8,8]
```

```python
plt.figure(figsize = (6,3))
plt.hist(data,color = "#D53487")
plt.xlabel("representing unique value")
plt.ylabel("representing frq")
plt.title("Hist plot example")
plt.grid()
plt.savefig("test.png")
# histogram means - how many times no are present in the given data
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
fig = plt.figure()
ax = fig.add_subplot(projection = "3d")
ax.scatter(x,y,z , c = "#9327E8",alpha = 0.9)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```
